```python
import requests
import pandas as pd
import os

url = "https://publicapi.traffy.in.th/ud/search"
limit = 999
offset = 0
results = []
```

```python
while True:
    # Make the request with the current offset
    response = requests.get(url, params={"offset": offset, "limit": limit})

    # Check if the results list in the response is empty
    if not response.json()['results']:
        break

    # Append the response results to the result list
    results.extend(response.json()['results'])

    # Increase the offset for the next request
    offset += limit
```

```python
# Create an empty DataFrame with the desired columns
columns = ['photo', 'coords', 'name', 'type', 'org', 'image_path']
df = pd.DataFrame(columns=columns)

# Loop through each item in the results list
for item in results:
    # Create the directory to save the image
    directory = "train_clip"
    if not os.path.exists(directory):
        os.makedirs(directory)

    # Extract the filename from the photo URL
    filename = item['photo'].split('/')[-1]

    # Check if the image already exists in the train_clip directory
    image_path = f"{directory}/{filename}"
    if not os.path.exists(image_path):
        # If the image does not exist, download it from the URL and save it
        response = requests.get(item['photo'])
        with open(image_path, 'wb') as f:
            f.write(response.content)

    # Add the image_path field to the item dictionary
    item['image_path'] = image_path

    # Append the item dictionary to the DataFrame
    df = df.append(item, ignore_index=True)

# Save the DataFrame to a CSV file
df.to_csv('dataset.csv', index=False)
```
